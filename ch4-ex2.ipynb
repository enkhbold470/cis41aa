```python
import random

# Inky Ganbold
# CIS41A
# Chapter 4 Exercise 2
# so this is for just printing some values, and getting the correct value until user gives 

ALLOWED_HOURS = 40
MUTLIPLICATION = 1.5

company = input("Enter your company name: ") #just getting the company name

while True: # for running it for infinite times until it reaches to the requested 
    hours_input = input("Enter the hours: ")
    try:
        hours = float(hours_input)
        if hours < 0:
            print("Error, please enter a positive number for hours")
            continue
        break
    except ValueError:
        print("Error, please enter numeric input")

while True:  # for running it for infinite times until it reaches to the requested 
    rate_input = input("Enter the rate: ")
    try:
        rate = float(rate_input)
        if rate < 0:
            print("Error, please enter a positive number for rate")
            continue 
        break
    except ValueError:
        print("Error, please enter numeric input") 

if hours > ALLOWED_HOURS: # this is just finding if user worked like butt off
    overtime_pay = (hours - ALLOWED_HOURS) * (rate * MUTLIPLICATION)
    total_pay = overtime_pay + (rate * ALLOWED_HOURS)
else:
    total_pay = rate * hours

doc_number = random.randint(1000, 2000)

print(f"\nCompany: {company}")
print(f"hours: {round(hours,2)}")
print(f"Rate: {round(rate,2)}")
print("**********************************************")
print(f"Your document number is: {doc_number}")
print(f"Your {company} gross pay is {round(total_pay,2)} dollars.")

'''
Test 1:
Enter your company name: Google
Enter the hours: 45
Enter the rate: 10

Company: Google
hours: 45.0
Rate: 10.0
**********************************************
Your document number is: 1585
Your Google gross pay is 475.0 dollars.

Test 2:
Enter your company name: Amazon
Enter the hours: forty
Error, please enter numeric input
Enter the hours: 40
Enter the rate: 15

Company: Amazon
hours: 40.0
Rate: 15.0
**********************************************
Your document number is: 1234
Your Amazon gross pay is 600.0 dollars.

Test 3:
Enter your company name: Apple
Enter the hours: -5
Error, please enter a positive number for hours
Enter the hours: 20
Enter the rate: nine
Error, please enter numeric input
Enter the rate: 20

Company: Apple
hours: 20.0
Rate: 20.0
**********************************************
Your document number is: 1999123
Your Apple gross pay is 400.0 dollars.
'''
```
